```python
countries = df.country.value_counts().reset_index(name="cnt")
```

```python
countries
```

```python
countries_modified =pd.DataFrame()
for index, row in countries.iterrows():
    if row['cnt'] > 70:
        countries_modified.at[index, 'country'] = row['country']
    else:
        countries_modified.at[index, 'country'] = 'other'
    countries_modified.at[index, 'cnt'] = round(int(row['cnt']),0)
```

```python
country_modified = countries_modified.groupby('country')['cnt'].sum().reset_index(name='number')
country_modified['total_count'] = country_modified['number'].sum()
country_modified['perc'] = round(country_modified['number']/country_modified['total_count']*100,1)
country_modified
```

```python
plt.figure(figsize=(10, 10)) 
plt.pie(country_modified.perc, labels=country_modified.country, labeldistance=1.0, textprops={'rotation': 42}, autopct = '%1.1f%%' )
plt.show()
```

```python
df['activity'] = df['activity'].str.strip()
df.activity = df.activity.apply(lambda sex:str(sex))
df['activity'] = df['activity'].str.lower()
```

```python
df.activity.value_counts().head(20)
```

```python
for i, value in enumerate(df['activity']):
    x = re.search(r'^.*surf.*$|^.*boarding.*$|^.*skiin.*$', value)
    y = re.search(r'^.*diving.*$', value)
    z = re.search(r'^.*fishing.*$', value)
    q = re.search(r'^.*swimming.*$|^.*snorkeling.*$|^.*treading.*$|^.*floating.*$', value)
    bath = re.search(r'^.*bathing.*$', value)
    stand = re.search(r'^.*standing.*$|^.*wading.*$|^.*walking.*$', value)
    row = re.search(r'^.*row.*$|^.*canoe.*$|^.*kayak.*$', value)
    if x:
        df.at[i, 'activity_masked'] = 'surfing/boarding-related'
    elif y:
        df.at[i, 'activity_masked'] = 'diving-related'
    elif z:
        df.at[i, 'activity_masked'] = 'fishing-related'
    elif q:
        df.at[i, 'activity_masked'] = 'swimming-related'
    elif bath:
        df.at[i, 'activity_masked'] = 'bathing-related'
    elif stand:
        df.at[i, 'activity_masked'] = 'standing/walking through the water'
    elif row:
        df.at[i, 'activity_masked'] = 'rowing-related'     
    else:
        df.at[i, 'activity_masked'] = 'other/unknown'     
```

```python
activities = df.activity_masked.value_counts().reset_index(name='cnt')
activities
```

```python
type(activities.cnt[0])
```

```python
plt.pie(activities.cnt, labels=activities.activity_masked, labeldistance=1.2, autopct='%1.1f%%')
plt.show()
```

```python
activity_sex = df.groupby(['activity_masked','sex']).date.count().reset_index(name='cnt')
```

```python
act_sex_cnt_total = activity_sex.groupby(['activity_masked']).cnt.sum().reset_index(name='cnt_total')
```

```python
act_sex_cnt_total
```

```python
merged_act_sex = pd.merge(activity_sex, act_sex_cnt_total, on='activity_masked', how='inner')
merged_act_sex['perc'] = round(merged_act_sex['cnt']/merged_act_sex['cnt_total']*100,1)
merged_act_sex
```

```python
import seaborn as sns
sns.barplot(x='activity_masked', y='perc', hue='sex', data=merged_act_sex)
plt.xticks(rotation=45, ha='right')
plt.show()
```
